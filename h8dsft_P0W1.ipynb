# Menangani missing Value

Sebelum data diubah, alangkah lebih baiknya kita membuat salinan data agar dapat membandingkan data original dengan data yang sudah kita rubah

# Membuat salinan data original

```python
pokemon = df.copy()
```

```python
pokemon.info()
# terdapat missing value pada data Type 2 yaitu berisikan 414 data saja dan akan dihapus agar tidak ada missing value dalam dataset
```

# Fillna

```python
pokemon['Type 2'].fillna('None', inplace=True) 
# data yang tidak ada isi akan di isi dengan 'None' atau tidak ada
```

```python
pokemon.info()
```

```python
pokemon.head(10)
```

# Manipulating Columns

```python
pokemon.rename(
    columns={'#' : 'No', 'Type 1' : 'Type_1', 'Type 2' :'Type_2'} , inplace=True
)
# Pergantian nama kolom # menjadi No, Type 1 menjadi Type_1, Type 2 menjadi Type_2
```

```python
pokemon.head()
```

# Query data

```python
pokemon[
    (pokemon['Type_2'] == 'Bug') &
    (pokemon['Speed'] > 30) &
    (pokemon['Attack'] > 100) &
    (pokemon['Type_1'] == 'Rock')
]
# pokemon dengan type 2 jenis bug, memiliki speed besar dari 30, attack besar dari 100 dan type 1 rock adalah Armaldo
```

```python
pokemon[
    (pokemon['Legendary'] == True) &
    (pokemon['Attack'] > 100) &
    (pokemon['Defense'] > 100) &
    (pokemon['Speed'] > 100) &
    (pokemon['Sp. Atk'] > 100) &
    (pokemon['Sp. Def'] > 100) 
]
# pokemon dengan jenis legendary yang memiliki attack, deff, sp.atk, sp.deff besar dari 100 adalah :
```

```python
pokemon[
    (pokemon['Legendary'] == False) &
    (pokemon['Attack'] > 100) &
    (pokemon['Defense'] > 100) &
    (pokemon['Speed'] > 100) &
    (pokemon['Sp. Atk'] > 100) &
    (pokemon['Sp. Def'] > 100) 
]
# pokemon dengan jenis tidak legendary yang memiliki attack, deff, sp.atk, sp.deff besar dari 100 adalah :
```

## Grouping dan aggregating

```python
pokemon.groupby('Type_1')['Attack'].max()
# ngegroup type 1 dengan attack tertinggi
```

```python
pokemon[
    (pokemon['Type_1'] == 'Psychic') &
    (pokemon['Attack'] > 140) &
    (pokemon['Speed'] > 120) 
].groupby(['Name', 'Attack','Speed', 'Sp. Atk', 'Sp. Def'])['Defense'].max()
# pokemon Type_1 Pyschic yang memiliki Attack besar dari 140, Speed besar dari 120, digroup dengan keterangan yang ditampilkan berdasarkan 
# Name, Attack, Speed, Sp. Atk, Sp. Def, Def, tertinggi
```

```python
pokemon[
    (pokemon['Type_1'] == 'Normal') &
    (pokemon['Attack'] < 10) 
].groupby(['Name','Attack', 'Speed', 'Sp. Atk', 'Sp. Def'])['Defense'].min()
# pokemon type_1 normal yang memiliki attack kecil dari 10, digroup dengan keterangan yang ditampilkan berdasarkan 
# att, spd, sp.att, sp.def, def, terendah
```
